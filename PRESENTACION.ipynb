## Introducción

This report explores the Copeland & Galai information-based model, which demonstrates how, even without transaction costs, a spread exists between the bid and ask prices in a financial market. This spread arises due to information asymmetry among market participants, where some (informed traders) possess private information about the true value of an asset, while others (uninformed or liquidity traders) trade for liquidity or other needs.

## Objetives

Implement the Copeland & Galai model in Python.
Analyze the impact of the proportion of informed traders (П₁) and the sensitivity of liquidity traders to prices (ПLB(S) and ILS(S)) on the market maker's revenue.
Determine the optimal bid and ask prices that maximize the market maker's revenue.

## Methodology

### Price distribution 

The price distribution is simulated using a Weibull distribution with parameters λ = 50 and K = 10.

### Expected revenue 

The market maker's expected revenue is calculated under three scenarios:

All traders are liquidity traders (П₁ = 0).

There is a 40% chance that an informed trader will execute a trade (П₁ = 0.4), keeping the same liquidity probabilities as in the previous scenario.

Consider П₁ = 0.4, ПLB(S) = 0.5 - 0.08S, and ILS(S) = 0.5 - 0.08S, where ПLB and ILS vary between 0 and 0.5.

### Price optimization

The Copeland & Galai model is used to determine the optimal bid and ask prices that maximize the market maker's revenue, considering the proportion of informed traders and the sensitivity of liquidity traders to prices.
